```python
import matplotlib.cm as cm #colormap
N = 256
angle = np.linspace(0, 8*2*np.pi, N)
radius = np.linspace(5, 1, N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X,Y, c=angle, cmap=cm.Spectral) #Greys, hsv, Reds, Blues, Greens
plt.show()
```

```python
X = np.linspace(-10, 10 ,100)
Y1 = [2*x**2+1 for x in X]
Y2 = [3*x**2+5 for x in X]
Y3 = [0.5*x**2+3 for x in X]
plt.plot(X, Y1, linestyle="solid", color='Red')
plt.plot(X, Y2, linestyle='dashed', color='blue')
plt.plot(X, Y3, linestyle='dashdot', color='green')
plt.show()
```

```python
N = 4
A = np.random.random(N)
B = np.random.random(N)
X = np.arange(N)
plt.bar(X, A, color = 'Red')
plt.bar(X, A+B, bottom = A)
```

```python
N = 4
A = np.array([1, 2, 3, 2])
B = np.array([2, 3, 2, 1])
C = np.array([1, 4, 2, 3])
X = np.arange(N)
plt.bar(X, A, color = 'Red')
plt.bar(X, A+B, color = 'Blue', bottom =A)
plt.bar(X, A+B+C, color = "green", bottom =B+A)
#plt.bar(X, A+B, color = 'green', bottom = A)
plt.show()
```

```python
#hatch
N = 4
A = np.array([1, 2, 3, 2])
B = np.array([2, 3, 2, 1])
C = np.array([1, 4, 2, 3])
X = np.arange(N)
plt.bar(X, A, color = 'Red', hatch='/')
plt.bar(X, A+B, color = 'Blue', hatch='x', bottom =A)
plt.bar(X, A+B+C, color = "green", hatch='y', bottom =B+A)
#plt.bar(X, A+B, color = 'green', bottom = A)
plt.show()
```

```python
import matplotlib.patches as patches
ax = plt.axes(polar=True)

radius = [1, 0.5, 0.7, 0.3, 0.1, 0.2, 0.6, 0.8, 0.5]
theta = np.linspace(0, 2*np.pi, len(radius), endpoint=False)

points = np.vstack((theta, radius)).T
plt.gca().add_patch(patches.Polygon(points, color = "0.5"))
plt.show()
```
