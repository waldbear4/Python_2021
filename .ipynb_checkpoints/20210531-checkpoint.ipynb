```python
# 라이브러리 로딩
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [1, 3, 4]
plt.plot(x ,y)
```

```python
x = [1, 3, 5]
y = [1, 4, 5]
plt.plot(x ,y, 'o--') # 점선처리
plt.show()
```

```python
X = range(-20, 21)
Y = [2*x+1 for x in X]
plt.plot(X, Y)
plt.show()
```

```python
X = range(0, 5) # 0에서 5까지
Y = [2*x+1 for x in X] # 선형회귀 
plt.plot(X, Y)
DataX = [1, 3, 4]
DataY = [2.9, 7.1, 9.3 ]
plt.plot(DataX, DataY, 'o')
plt.show()
```

```python
import math as m # y = sin(2*pi*t)
X = range(0, 20) # 그래프의 범위
Y = [m.sin(x) for x in X] # 선형회귀 
plt.plot(X, Y, 'o--')
plt.show()
```

```python
X = np.linspace(0, 1, 10) # 0 에서 1 까지 10개를 만들어보자
print(X)
```

```python
X = np.linspace(0, 2*np.pi, 100) #0에서 np.pi 가져와서 100등분
Y = np.sin(X)
plt.plot(X, Y)
plt. show()
```

```python
# scatter
data = np.random.rand(100, 2) #100행 2열 X,Y
data.shape
```

```python
plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

```python
data1 = np.random.rand(200, 2)
data2 = np.random.rand(200, 2)
data3 = np.random.rand(200, 2)
data1 += np.array((-2, -2))
data2 += np.array((2, 0))
data3 += np.array((2, -2))
plt.scatter(data1[:, 0], data1[:, 1])
plt.scatter(data2[:, 0], data2[:, 1])
plt.scatter(data3[:, 0], data3[:, 1])
plt.show()
```

```python
data = [5, 25, 100, 50]
index = [1, 2, 3, 4]
plt.bar(index, data, color = "b",)
plt.show()
```

```python
data = [5, 25, 100, 50]
index = [10, 20, 30, 40]
plt.bar(index, data, color = "b",)
plt.show()
```

```python
data = [5, 25, 100, 50]
index = ["a", "b", "c", "d"]
plt.bar(index, data, width=1 ,color = "b",)
plt.show()
```

```python
data = [5, 25, 100, 50]
index = ["a", "b", "c", "d"]
plt.barh(index, data, color = "b",)
plt.show()
```

```python
data1 = [5, 20, 60, 10];index1 = [1, 6, 11, 16]
data2 = [10, 25, 100, 50];index2 = [2, 7, 12, 17]
data3 = [20, 50, 30, 50];index3 = [3, 8, 13, 18]
plt.bar(index1, data1, color = "b",)
plt.bar(index2, data2, color = "r",)
plt.bar(index3, data3, color = "g",)

plt.show()
```

```python
dataA = np.random.standard_normal((100, 2)) #평면에 100개의 점을 찍는다
dataB = np.random.standard_normal((100, 2))
dataC = np.random.standard_normal((100, 2))
dataA += np.array((-3, -3))
dataB += np.array((4, 4))
dataC += np.array((-1, 4))
plt.scatter(dataA[:,0],dataA[:,1], color="k")
plt.scatter(dataB[:,0],dataB[:,1], color="0.6")
plt.scatter(dataC[:,0],dataC[:,1], color="#79AAFF")
plt.show()
```
