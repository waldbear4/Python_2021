KMenas 는 클러스터 간에 균일하게 가져가려는 성향
갯수가 중심점을 중심으로 거리로 계산해서
밑의 파란 점들을 보면 분포를 확인하고
위의 라벨링된 그래프를 보면 

비지도 학습으로 먼저 성향과 경향을 보고 
지도 학습으로 포괄적인 부분을 수정함
딥러닝으로 세부적인 부분을 나눔
ex)X-Ray 사진에는 명약관화한 종양부분이 보이지만, 실질적으로는 다른 작은 종양이 실제할 수 있음.
그렇다면 수 많은 엑스레인 데이터가 있으면 각자의 부분을 비교하여 진단율을 높일 수 있지 않을까?
적층방식, 레이어를 겹쳐서 수 많은 부분에 있어서 비교 데이터를 가지고 있으면 y 라는 파라메터를 찾을 수 있지 않을까.

```python
X, y = make_blobs(random_state=170, n_samples=600)
rng = np.random.RandomState(74)
```

무작위로 클러스터 데이터 생성

```python
transfomation = rng.normal(size=(2, 2))
X = np.dot(X, transfomation)
```

데이터가 길게 늘어지도록 변경

```python
kmaens = KMeans(n_clusters=3).fit(X)
y_pred = kmaens.predict(X)
```

세 개의 클러서트로 데이터에 KMeans 알고리즘을 적용

```python
mglearn.discrete_scatter(X[:,0], X[:,1], markers='o')
```

라벨제거

```python
mglearn.discrete_scatter(X[:,0], X[:,1],y, markers='o')
```

y 로 라벨 설정

```python
mglearn.discrete_scatter(X[:,0], X[:,1],y_pred, markers='o')
```

y_pred 라벨

```python
mglearn.discrete_scatter(X[:,0], X[:,1], kmaens.labels_, markers='o')
mglearn.discrete_scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], [0,1,2],
                         markers=' ^', markeredgewidth=2)
plt.xlabel("C0")
plt.ylabel("C1")
```

클러스 할당과 클러스 중심을 나타냄 + kmeanslabel 설정

즉 kmeans는 중심점을 중심으로 거리로만 클러스터를 카테고리하기 때문에, 실제 데이터셋의 경향을 무시하고 그룹핑을함

```python
mglearn.discrete_scatter(X[:,0], X[:,1], kmaens.labels_, markers='o')
mglearn.discrete_scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], [0,1,2],
                         markers=' ^', markeredgewidth=2)
mglearn.discrete_scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], markers='s')
plt.xlabel("C0")
plt.ylabel("C1")
```

```python
#p233
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise = 0.05, random_state=0)

model = KMeans(n_clusters=2).fit(X)
y_ = model.predict(X)

mglearn.discrete_scatter(X[:,0], X[:,1], y)
```

일반 데이터 셋

```python
mglearn.discrete_scatter(X[:,0], X[:,1], y_)
```

KMeans 의 결과

```python
mglearn.discrete_scatter(X[:,0], X[:,1], y_)
mglearn.discrete_scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], markers='s')
```
