```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

KMeans = K 갯수의 평균

```python
X, y = make_blobs(random_state = 1)
```

인위적인 2차원 데이터 생성

```python
mglearn.discrete_scatter(X[:,0], X[:,1])
```

Y를 주지 않았기 때문에 색상으로 구분이 안 됨

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
X
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
kmeans
```

K개의 군집을 준다 > n_cluster

```python
X, y = make_blobs(random_state = 1)
kmeans = KMeans(n_clusters=3).fit(X)
for i in range(len(y)):
    print(kmeans.labels_, y)
```

```python
mglearn.discrete_scatter(X[:,0], X[:,1], y)
```

스스로 만들어 낸 라벨을 출력하라

```python
mglearn.discrete_scatter(X[:,0], X[:,1], kmeans.labels_)
```

```python
X, y = make_blobs(random_state = 1, cluster_std= 3)
mglearn.discrete_scatter(X[:,0], X[:,1], kmeans.labels_)
```

```python
#p230
X_varied, y_varied= make_blobs(n_samples=200, random_state = 170, cluster_std= [1, 2.5, 0.5,])
y_pred = KMeans(n_clusters=3, random_state=0).fit_predict(X_varied)
mglearn.discrete_scatter(X_varied[:,0], X_varied[:,1], y_pred)
plt.legend(["C0", "C1", "C2"], loc="best")
plt.xlabel("Character 0")
plt.ylabel("Character 1")
```
